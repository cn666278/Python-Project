```python
# Exchange sort

import random
import math
import time
ID=[]
cost=[]
Order={}
order={}

def exchange_sort(arr,cs):
    if len(arr) == 1:
        return arr,cs
    for i in range(1,len(arr)):
        for j in range (0,len(arr)-i):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                cs[j],cs[j+1]=cs[j+1],cs[j]
    return arr,cs

def zuzhuang(x,y,square):
    z={}
    for i in range(len(x)):
        id = "FD" + str(x[i]).zfill(square)
        Cost = "RM" + str(round(y[i],2))
        z[id] = Cost
    return z

def display(Order):
    for c in Order:
        print("%-5s   %5s" %(c,Order[c]))
    print("\n")

def main():
    
    print("Welcome to the Order management system\n")

    n=int(input("Please input the maximum amount of n:"))
    
    
    # Random generate ID
    ID = random.sample(range(1,n+1),n)
    
    # Random generate cost
    for i in range(0,n):
        cost.append(random.uniform(100,1000))

    # Put it in the dictionary(Unsorted)
    cifang = math.ceil(math.log(n,10))+1
    order = zuzhuang(ID,cost,cifang)    
    exchange_sort(ID,cost)
    
    choice = 0
    while(choice != -1):
        print("【1】Time taken to perform the sorting")
        print("【2】All the unsorted order ids and the corresponding total order cost that were generated")
        print("【3】All the sorted order ids and the corresponding total order cost that were generated")
        print("【-1】Exit")
        choice = int(input())
        if choice == 1:
            %timeit time_list = exchange_sort(ID,cost)
            print("\n")
        elif choice == 2:
            display(order)
        elif choice == 3:
            # Put it in the dictionary(Sorted)
            Order = zuzhuang(ID,cost,cifang)
            display(Order)

main()
```

```python
# Radix Sort
import random
import time

def generateID(n):
    global ids 
    ids = {}
    global arr
    arr = random.sample(range(1,n+1),n)
    for i in range(0,len(arr)):
        ids["FD"+str(arr[i])] = "RM" + str(random.randint(0, 99999)/100)
        
def generateSortedID(arr):
    global sortedids
    sortedids = {}
    for i in range(0,len(arr)):
        sortedid = "FD" + str(arr[i])
        sortedids["FD" + str(arr[i])] = ids.get(sortedid)

def radixSort(arr):
    i = 0 
    max_num = max(arr)  
    j = len(str(max_num))  
    while i < j:
        bucket_list =[[] for _ in range(10)] 
        for x in arr:
            bucket_list[int(x / (10**i)) % 10].append(x) 
        arr.clear()
        for x in bucket_list:   
            for y in x:
                arr.append(y)
        i += 1


n = int(input("please enter the number of ids "))
generateID(n)
start = time.time()
radixSort(arr)
end = time.time()
usedTime = end - start
print("The running time is " , usedTime)
generateSortedID(arr)
%timeit time_list = radixSort(arr)
choose = 1

while choose != 0:
    print("\nenter 1 to see the unsorted id\nenter 2 to see sorted id\nenter 0 to exit\n")
    choose = int(input("\n"))
    if choose == 1:
        print(ids)
    elif choose == 2:
        print(sortedids)
```
