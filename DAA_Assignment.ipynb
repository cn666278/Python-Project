```python
# Exchange sort

import random
import math
import time
ID=[]
cost=[]
Order={}
order={}

def exchange_sort(arr,cs):
    if len(arr) == 1:
        return arr,cs
    for i in range(1,len(arr)):
        for j in range (0,len(arr)-i):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                cs[j],cs[j+1]=cs[j+1],cs[j]
    return arr,cs

def zuzhuang(x,y,square):
    z={}
    for i in range(len(x)):
        id = "FD" + str(x[i]).zfill(square)
        Cost = "RM" + str(round(y[i],2))
        z[id] = Cost
    return z

def display(Order):
    for c in Order:
        print("%-5s   %5s" %(c,Order[c]))
    print("\n")

def main():
    
    print("Welcome to the Order management system\n")

    n=int(input("Please input the maximum amount of n:"))
    
    
    # Random generate ID
    ID = random.sample(range(1,n+1),n)
    
    # Random generate cost
    for i in range(0,n):
        cost.append(random.uniform(100,1000))

    # Put it in the dictionary(Unsorted)
    cifang = math.ceil(math.log(n,10))+1
    order = zuzhuang(ID,cost,cifang)    
    exchange_sort(ID,cost)
    
    choice = 0
    while(choice != -1):
        print("【1】Time taken to perform the sorting")
        print("【2】All the unsorted order ids and the corresponding total order cost that were generated")
        print("【3】All the sorted order ids and the corresponding total order cost that were generated")
        print("【-1】Exit")
        choice = int(input())
        if choice == 1:
            %timeit time_list = exchange_sort(ID,cost)
            print("\n")
        elif choice == 2:
            display(order)
        elif choice == 3:
            # Put it in the dictionary(Sorted)
            Order = zuzhuang(ID,cost,cifang)
            display(Order)

main()
```

# Merge sort
***

## Time complexity:


- ### Best case:

    The best case is the last comparison of an ordered subsequence with the last data element left, and another ordered subsequence with n/2 elements left, C(n) = n/2.  


> B(n) = B(⌊ n/2 ⌋) + B(⌈ n/2 ⌉) + (n/2) = 2B(n/2) + (n/2)  

> By the master method case 2, we have f(n) = (1/2)n ∈ Ꙫ(n) = Ꙫ(n^lg2)  
=>    W(n) ∈ Ꙫ(nlgn)
  

- ### Worst case:

    The worst case is the last comparison between two ordered subsequences, each with the last data element left. For example, the combination of these two subsequences takes n - 1 comparisons, which is the worst case, C(n) = n - 1.  


> W(n) = W(⌊ n/2 ⌋) + W(⌈ n/2 ⌉) + (n-1) = 2W(n/2) + (n-1)  

> By the master method case 2, we have f(n) = n ∈ Ꙫ(n) = Ꙫ(n^lg2)  
=>    W(n) ∈ Ꙫ(nlgn)
 

- ### Average case:

    The best case and the worst case are on the order of nlogn, so it's easy to conclude (the squeeze criterion of analogy and advanced mathematics) that the average merge sort is also on the order of nlogn.      
  
    If the time complexity of merge sort for n elements is T(n), then since merge sort uses the divide-and-conquer idea, T(n)=2 * T(n/2) + N, where N is the time complexity of merging two subintervals, as shown by the merge function. The following formula can be derived: 

> When  n = 1, only constant execution time is required, so it is denoted as C, then T(1) = C.  
When n > 1, T(n) = 2 * T(n/2) + n.  



And then we get T(n)=2^k * T(n/2^k) + k * n, and we assume that T(1) = T(n/2^k), which means that when we merge sort n/2^k elements, and we reach the recursive stop condition, n/2^k=1, we get: k=logn.  

Hence:  
T(n) = Cn + nlog2n  
The time complexity of merge sort is O(nlogn), which has nothing to do with the order of the array, also you can see from the graph, it's very stable.    
  
## Space complexity:
  The merge sort we used is not an in-place sort because it uses the array left and right bisides the input array.  
  And the total number of extra array items is: n + n/2 + n/4 + ... = 2n.

    Hence, the time complexity is O(n^2)

    The space complexity is O(n)

# Analyze Quicksort   

## Why we choose Quicksort rather than Radixsort?
Radix sort has the fastest speed in these five sort algorithms. And it's very steady in the best/worst/average case.

So why we not choose Radixsort?

When we are doing the Radixsort, we need to set auxiliary arrays in order to help sort. Hence, its space complexity is much higher than the Quicksort.

Compared to the small amount of time saved, resulting in a huge amount of space, this is very uneconomical

## The advantages of Quicksort
* In place swap, do not need extra space
* Time complexity is lower than most of sorting algorithms in average and best case.

## The disadvantages of Quicksort
* It's not steady, when quicksort in the worst case, it will degrade into exchange sort.

The three numbers I chosen are the arr[left] arr[mid] and arr[right]. Then choose the median of them as the pivot.

    def quicksort(arr,left,right):
        if left>=right:
            return
        else:
            mid = left + ((right - left)>1)
            if arr[mid]>arr[right]:
                swap(arr,mid,right)
            if arr[left] > arr[right]:
                swap(arr,left,right)
            if arr[mid] > arr[left]:
                swap(arr,mid,left)

            swap(arr,left,right)
            pivot = arr[right]
            i = left-1
            for j in range(left,right):
                if arr[j]<=pivot:
                    i = i+1
                    arr[i], arr[j]=arr[j], arr[i]
            arr[i+1],arr[right] = arr[right],arr[i+1]

            return i+1
    

    def swap(arr, i, j):
        arr[i], arr[j] = arr[j], arr[i]

    def insertionSort(arr):
        for i in range(len(arr)):
            preIndex = i-1
            current = arr[i]
            while preIndex >= 0 and arr[preIndex] > current:
                arr[preIndex+1] = arr[preIndex]
                preIndex-=1
            arr[preIndex+1] = current
    return arr
