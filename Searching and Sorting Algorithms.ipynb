```python
# Binary Search
def binary_search(S, low, high, key):
    if high >= low:
        mid = (low + high) // 2  # "//" means floor function
        
        if S[mid] == key:
            return mid
        elif S[mid] > key:
            return binary_search(S, low, mid - 1, key)
        else:
            return binary_search(S, mid + 1, high, key)
    else:
        return -1
    
    
S = []
n = int(input('Enter the number of elements in array:'))

for i in range(0,n):
    num_of_element = int(input('Array element %dï¼š ' %i))
    S.append(num_of_element)
    
key = int(input('Enter the key value to search:'))

output = binary_search(S, 0, len(S) - 1, key)

if output != -1:
    print('Element found at location: ', str(output))
else:
    print('Element is not found')
    
```

```python
# Merge Sort
def merge_sort(arr):
    if len(arr) == 1:
        return arr
    # using divide method to divide array into two sub array
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    # recursive left and right array
    return merge(merge_sort(left), merge_sort(right))


def merge(left, right):
    result = []
    # make sure the left and right array is not empty
    while len(left) > 0 and len(right) > 0:     
        # do comparison, the smaller one add to the font of result
        if left[0] <= right[0]:
            result.append(left.pop(0))
        else:
            result.append(right.pop(0))
    # if there still has values in the array(L/R), add them directly
    result += left
    result += right
    return result


arr=[1,2,6,8,5,10,7,22,3]
print(arr)
print(merge_sort(arr))
```

```python
# Quick Sort
import numpy as np

def quicksort(low, high, S):
    #     pivotPtr = S[low]
    if (high > low):
        pivotPtr = partition(low, high, S)
        quicksort(low, pivotPtr - 1, S)
        quicksort(pivotPtr + 1, high, S)
    return S


def partition(low, high, S):
    i = low + 1
    j = high
    pivot = S[low]

    while True:  # scan left and right, check whether scan is over and exchange the elements
        while S[i] < pivot: # when S[i] < pivot, let i=i+1
            i += 1
            if i == high:
                break
        while pivot < S[j]: # when pivot < S[j], let j=j-1
            j -= 1
            if j == low:
                break
        if i >= j:
            break
        S[i], S[j] = S[j], S[i]  # exchange S[j] and S[i] so that all element on left side of i is less than pivot

    S[low], S[j] = S[j], S[low] # when pointer i and j meet, exchange S[low] and S[j]
    return j


S = [1, 3, 6, 9, 2, 8, 7, 22, 4]
np.random.shuffle(S) # randomize the array
# print(S)
print(quicksort(0, len(S) - 1, S))
```

```python
# Heap Sort
def makeHeap(arr, n):
    for i in range(n//2 ,-1,-1):       # last parent should be n // 2 --(parent*2+1 is left chid)
        siftDown(arr, i, n)

def siftDown(arr, i, n): 
    parent = i  
    largerchild = i 
    l = 2 * parent + 1   # using arr[0] as root
    r = 2 * parent + 2   
    
    if l < n and arr[l] > arr[largerchild]:
        largerchild = l
        
    if r < n and arr[r] > arr[largerchild]: 
        largerchild = r

    # if the largest value changes(not parent node anymore)
    if largerchild != i:
        arr[i], arr[largerchild] = arr[largerchild], arr[i]    # update largest value 
        siftDown(arr, largerchild, n)                          # sink down

def heapSort(arr): 
    n = len(arr)
    makeHeap(arr,n)
    
    # exchange values one-by-one
    for i in range(n-1, 0, -1):           # start from n-1, means delete(ignore) the bottom node
        arr[i], arr[0] = arr[0], arr[i]    # exchange arr[0] with root(arr[n-1])
        siftDown(arr, 0, i)                # i = n-1,n-2 ...1, after excahnge arr[0] with root
                                           # sift down arr[i] to restore the heap property

        
arr = [12, 11, 1, 9, 13, 2, 8, 10, 5, 6, 7, 4, 3] 
# arr = ['a','s','f','g','z','d','c','e','n','o','b']
print(arr)
heapSort(arr) 
print ("After heap sort:") 
print(arr)
```
