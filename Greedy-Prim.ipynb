```python
# English Version
# Prim algorithm for minimum spanning tree using Greedy Approach

import numpy as np


def prim(n, W, F):
    global vnear

    for i in range(0, n):
        nearest[i] = 0         # for all vertices(v0-vn)，initialize v0 to be the nearest vertex in Y
        distance[i] = W[0][i]  # initialize the distance from Y to be the weight on the edge to v0

    for k in range(n - 1):  # Add all n-1 vertices to Y
        min = np.inf  # INF
        for i in range(1, n):  # Check each vertex (started from v1）for being nearest to Y
            if 0 <= distance[i] < min:
                min = distance[i]
                vnear = i  # vnear is the index of selected vertex to be added

        # e = edge connecting two vertices indexed by vnear and nearest[vnear]
        # add e to F
        F.append(W[vnear][nearest[vnear]])
        distance[vnear] = -1  # Add vertex indexed by vnear to Y

        for i in range(0, n):
            if W[i][vnear] < distance[i]:  # For each vertex not in Y, update its distance from Y
                distance[i] = W[i][vnear]
                nearest[i] = vnear
    return F


_ = np.inf
W = [  # the graph
    [0, 1, 3, _, _],  # 0 means（v0）, 1 means the distance of v0-v1 ...
    [1, 0, 3, 6, _],  # 0 means（v1）
    [3, 3, 0, 4, 2],
    [_, 6, 4, 0, 5],
    [_, _, 2, 5, 0],
]

Y = []  # selected vertexes,initially, Y={v0}
F = []  # selected edges
n = len(W[0])  # n>=2

# initialization(otherwise can not assign list[i])
nearest = [0] * n  # The index of the vertex that nearest to the vi inside Y
distance = [0] * n  # weight of edge，connecting between vi and vertex indexed by nearest[i] 
print(prim(n, W, F))
```
