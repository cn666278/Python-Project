```python
# Using patial list element(slice)
players = ['Charley','Pob','Gayer','Kanyer','Bli']
for player in players:
    print(player)
print("\n")
print(players[0:3]) # 0 1 2
print(players[1:4]) # 1 2 3 
print(players[2:5])# 2 3 4
print(players[:5]) # start from the head
print(players[2:]) # stop until see the tail
print(players[-2:]) # output the last 2 elements

# Traverse your slice
print("Here are the first 3 players in our team:")
for player in players[0:3]:
    print(player.title())

# Copy your list
myfood = ['Pizza','Hamberge','Meat']
friendfood = myfood[:]
myfood.append('Fish')
friendfood.append('Cake')
print('\nMy favourite food are:')
print(myfood)
print('\nMy friend favourint food are:')
print(friendfood)

# Practice
lists = ['GAY','BILI','FYUIX','PYNIX','FXXK']
print("\nThe first three items in the list are:")
for list in lists[0:3]:
    print(list)
print("\nThe 3 items from the middle of the list are：")
for list in lists[1:4]:
    print(list)
print("\nThe last 3 items from the list are:")
for list in lists[-3:]:
    print(list)
friend_pizzas = ['seafood','hotpot','cheese']
my_pizzas = friend_pizzas[:]
friend_pizzas.append("BBQ flavor")
my_pizzas.append("Meat")
print('\nMy favourite pizzas are:')
print(my_pizzas)
print('\nMy friend favourint pizzas are:')
print(friend_pizzas)
print("\n")
for pizza in my_pizzas:
    print('I like ' + pizza + " pizza!")
```
