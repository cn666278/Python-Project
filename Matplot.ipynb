```python
def my_plotter(ax, data1, data2, param_dict):
    #ax : Axes，The axes to draw to
    fig, ax = plt.subplots()
    ax.set_xlabel('Input size n')  # Add an x-label to the axes.
    ax.set_ylabel('Time')  # Add a y-label to the axes.
    ax.set_title("Compare all 5  sort algorithms")  # Add a title to the axes.
    ax.legend()  # Add a legend.
    
    #data1 : array，the x data       

    #data2 : array, the y data
       

    # param_dict : dict
    #   Dictionary of kwargs to pass to ax.plot

    #Returns
    
    plt.ylim((0,500)) # 限制y轴的值
    
    out = ax.plot(data1, data2, **param_dict)
    return out

data1 = [1000,5000,32000]
data2 = [44.3, 1300, 54500]
data3 = [3.64,21.6,157]
data4 = [6.03,40.5,333]
data5 = [4.81,31.4,358]
my_plotter(ax, data1, data2, {'marker': 'x'})
my_plotter(ax, data1, data3, {'marker': 'x'})
```
