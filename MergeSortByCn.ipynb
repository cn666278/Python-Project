## Merge sort

### Time complexity:


- #### Best case:
The best case is the last comparison of an ordered subsequence with the last data element left, and another ordered subsequence with n/2 elements left, C(n) = n/2.  

  
- #### Worst case:
The worst case is the last comparison between two ordered subsequences, each with the last data element left. For example, the combination of these two subsequences takes n - 1 comparisons, which is the worst case, C(n) = n - 1.  
  
  
- #### Average case:
The best case and the worst case are on the order of nlogn, so it's easy to conclude (the squeeze criterion of analogy and advanced mathematics) that the average merge sort is also on the order of nlogn.      
  
If the time complexity of merge sort for n elements is T(n), then since merge sort uses the divide-and-conquer idea, T(n)=2 * T(n/2) + N, where N is the time complexity of merging two subintervals, as shown by the merge function. The following formula can be derived: 

> When  n = 1, only constant execution time is required, so it is denoted as C, then T(1) = C.  
When n > 1, T(n) = 2 * T(n/2) + n.


And then we get T(n)=2^k * T(n/2^k) + k * n, and we assume that T(1) = T(n/2^k), which means that when we merge sort n/2^k elements, and we reach the recursive stop condition, n/2^k=1, we get: k=logn.  

Hence:  
T(n) = Cn + nlog2n  
The time complexity of merge sort is O(nlogn), which has nothing to do with the order of the array, also you can see from the graph, it's very stable.    
  
### Space complexity:
  The merge sort we used is not an in-place sort because it uses the array left and right bisides the input array.  
  And the total number of extra array items is: n + n/2 + n/4 + ... = 2n.
